### Divisi贸n de los datos

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size=0.8,
                                                    random_state=2026)

X_train.shape, X_test.shape
```

### Entrenamiento e Inferencia

```python
print(y_test[:5])
print(y_pred[:5])
```

## Evaluaci贸n

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='macro')
cm = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"F1: {f1}")
print(cm)
```

# Regresi贸n

```python
import pandas as pd

train_data_df = pd.read_csv('/content/sample_data/california_housing_train.csv')
test_data_df = pd.read_csv('/content/sample_data/california_housing_test.csv')

train_data_df.head()
# X.shape, y.shape
```

```python
X_train = train_data_df.drop(columns=['median_house_value'])
y_train = train_data_df['median_house_value']

X_test = test_data_df.drop(columns=['median_house_value'])
y_test = test_data_df['median_house_value']

X_train.shape, X_test.shape
```

## Entrenamiento e Inferencia

Tambi茅n podemos usar una red neuronal

```python
#---- ESCOGEMOS EL MODELO DE REGRESIN ----
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor

#--- INICIALIZAMOS EL MODELO Y LO ENTRENAMOS ---
regresor = LinearRegression()
# regresor = SVR()
# regresor = DecisionTreeRegressor()

regresor.fit(X_train, y_train)

#--- INFERENCIA ---
y_pred = regresor.predict(X_test)
```

## Evaluaci贸n

```python
#--- ESCOGEMOS LA MTRICA DE EVALUACIN DE LA REGRESIN ---
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

#--- EVALUAMOS EL MODELO ---
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)

print(f"MSE: {mse}")
print(f"MAE: {mae}")
print(f"MAPE: {mape}")
```

# Clustering

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=1000, n_features=2, centers=4, cluster_std=0.25)

plt.figure()
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
X.shape
```

## Entrenamiento e Inferencia

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN

clustering = KMeans(n_clusters=4)
# clustering = AgglomerativeClustering(n_clusters=4)
# clustering = DBSCAN(eps=0.1)
clustering.fit(X)
```

```python
labels = clustering.labels_
labels[:10]
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```

## Evaluaci贸n

```python
#--- ESCOGEMOS LA MTRICA DE EVALUACIN DE LA REGRESIN ---
from sklearn.metrics import silhouette_score

sil = silhouette_score(X, labels)
print(f"Score de silueta: {sil}")
```
