Leemos el archivo `csv` con [pandas](https://pandas.pydata.org/docs/user_guide/index.html) en un [dataframe](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html#pandas.DataFrame)

```python
import pandas as pd

df = pd.read_csv('IMDB.csv')
df
```

Podemos ver que es un dataset etiquetado

```python
import wordcloud
import matplotlib.pyplot as plt

text = " ".join(text for text in df['review'])

wc = wordcloud.WordCloud(background_color="white").generate(text)

plt.figure()
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

veamos por etiquetas:

```python
import wordcloud
import matplotlib.pyplot as plt

pos_df = df[df['sentiment']=='positive'].copy()
neg_df = df[df['sentiment']=='negative'].copy()
dfs = [pos_df, neg_df]

for x,label in zip(dfs,['positivos','negativos']):
    text = " ".join(text for text in x['review'])
    wc = wordcloud.WordCloud(background_color="white").generate(text)
    plt.figure()
    plt.suptitle(label)
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    plt.tight_layout()
    plt.show()
```
